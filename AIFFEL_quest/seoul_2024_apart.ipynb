```python
unique_nan_addresses = df_test['주소'].unique()
location_dict = {}

def get_geocode(address):
    try:
        geo_loc = maps.geocode(address)
        lat = geo_loc[0]['geometry']['location']['lat']
        lng = geo_loc[0]['geometry']['location']['lng']
        return [lat,lng]
    except:
        return [None, None]

for address in unique_nan_addresses:
    location = get_geocode(address)
    location_dict[address] = (location[0], location[1])
```
