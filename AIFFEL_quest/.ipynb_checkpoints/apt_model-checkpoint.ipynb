```python
def haversine(lon1, lat1, lon2, lat2):
    R = 6371  # 지구의 반지름 (킬로미터 단위)
    dlon = np.radians(lon2 - lon1)
    dlat = np.radians(lat2 - lat1)
    a = np.sin(dlat / 2) ** 2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# 아파트별로 가장 가까운 지하철역 찾기
def find_nearest_station(apartments, j, k):
    
    index = np.where(df['주소'] == apartments)[0]
    
    apt_lat, apt_lon = df['위도'].iloc()[index], df['경도'].iloc()[index]
    station_lat, station_lon = j, k
    dist = haversine(apt_lon, apt_lat, station_lon, station_lat)

    return distances
```
