```python
with open('/content/drive/MyDrive/Colab Notebooks/Avengers.txt', 'rt') as text:
  text = text.read()
  text = text.lower()
  text = text.replace('\n', '')
  text = text.replace(',', '')
  text = text.replace('.', '')
  text = text.replace('?', '')
  text = text.replace('!', '')
  text = text.replace('"', '')
  text = text.replace("'", '')
  text = text.replace('(', '')
  text = text.replace(')', '')
  text = text.replace('-', '')
  text = text.split(" ")
  print(text)

  gen = [text[i] + text[i + 1] for i in range(len(text) - 1)]
  count = 0
  diction = {}
  for g in gen:
    for i in gen:
      if g == i:
        count += 1
    diction[g] = count
    print(g + " = " + str(count))
    count = 0
  print(max(diction.values))
```
